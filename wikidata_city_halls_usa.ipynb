```python
import requests
from bs4 import BeautifulSoup as BS
import urllib.request
```

This query string queries every building that `is an instance of` `city hall` with `country` `United States of America`.

```python
url = 'https://query.wikidata.org/sparql'
query = """
SELECT ?building ?buildingLabel
WHERE
{
  ?building wdt:P31 wd:Q543654;
            wdt:P17 wd:Q30.  
  SERVICE wikibase:label { bd:serviceParam wikibase:language "[AUTO_LANGUAGE]".}
}
"""
```

```python
r = requests.get(url, params = {'format': 'json', 'query': query})
data = r.json()
bindings = data['results']['bindings']
qnums = [binding['buildingLabel']['value'] for binding in bindings]
```

```python
len(qnums)
```

```python
titles = []
for ind in range(len(qnums)):
    if ind % 10 == 0:
        print(ind)
    qnum = qnums[ind]
    url ="https://www.wikidata.org/wiki/%s" % qnum
    usock = urllib.request.urlopen(url)
    data = usock.read()
    usock.close()
    soup = BS(data)
    titles.append(soup.find('title').text)
```

```python
names = [title[:-11] for title in titles]
```
